```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-10,10,1000)
y = np.sign(x)/(1+abs(x))

plt.figure(figsize=(10,6))
plt.plot(x,y)
plt.xlabel('Integration (Control)',fontsize=24)
plt.ylabel('Integration (RL)',fontsize=24)
plt.title('Map the control integration into RL observation',fontsize=24)
plt.grid()

plt.savefig('integration')
```

```python
x = np.linspace(-10,10,1000)
y = np.sign(x)*abs(x)/(1+abs(x))

plt.figure(figsize=(10,6))
plt.plot(x,y)
plt.xlabel('Integration (Control)',fontsize=24)
plt.ylabel('Integration (RL)',fontsize=24)
plt.title('Map the control integration into RL observation',fontsize=24)
plt.grid()

plt.savefig('integration_correct')
```
